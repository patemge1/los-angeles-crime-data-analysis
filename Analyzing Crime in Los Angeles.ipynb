# Los Angeles Crime Data Analysis

## Overview
This project analyzes crime data from Los Angeles to identify patterns in criminal activity across time, location, and victim demographics. The goal is to uncover insights that can support data-driven decision-making for resource allocation and operational planning.

---

## Dataset

The dataset is a modified version of publicly available data from Los Angeles Open Data. It contains detailed records of reported crimes, including temporal, geographic, and demographic attributes.

### File: `crimes.csv`

| Column | Description |
|--------|-------------|
| `DR_NO` | Unique record identifier composed of year, area ID, and sequence number |
| `Date Rptd` | Date the crime was reported |
| `DATE OCC` | Date the crime occurred |
| `TIME OCC` | Time of occurrence (24-hour format) |
| `AREA NAME` | Geographic patrol division responsible for the area |
| `Crm Cd Desc` | Description of the crime committed |
| `Vict Age` | Victim's age |
| `Vict Sex` | Victim's sex (F, M, X) |
| `Vict Descent` | Victim's descent classification |
| `Weapon Desc` | Description of weapon used (if applicable) |
| `Status Desc` | Status of the crime case |
| `LOCATION` | Street address of the incident |

---

## Analysis Focus

The analysis focuses on identifying:
- Peak crime hours
- Areas with high nighttime crime activity
- Crime distribution across victim age groups

These insights provide a foundation for understanding patterns in criminal behavior and identifying potential areas for operational improvement.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", dtype={"TIME OCC": str})
crimes.head()
```

```python
# ---------------------------------------------
# 1. Analyze crime frequency by hour of occurrence
# ---------------------------------------------
# Extract hour from TIME OCC (first two characters) and convert to integer
crimes["HOUR"] = crimes["TIME OCC"].str[:2].astype(int)

# Identify the most frequent hour for crimes
peak_crime_hour =  crimes["HOUR"].mode()[0]

print (peak_crime_hour)
```

```python
# ---------------------------------------------
# 2. Identify area with highest night crime activity
# ---------------------------------------------

# Filter dataset for nighttime hours (10 PM – 3 AM)
night_crimes = crimes[crimes["HOUR"].isin([22, 23, 0, 1, 2, 3])]

# Determine the area with the highest frequency of nighttime crimes
peak_night_crime_location = night_crimes ["AREA NAME"].mode()[0]

print (peak_night_crime_location)
```

```python
# ---------------------------------------------
# 3. Analyze crime distribution by victim age group
# ---------------------------------------------

# Define age group bins and corresponding labels
bins = [0, 17, 25, 34, 44, 54, 64, np.inf]
labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]

# Categorize victim ages into defined groups
crimes["age_group"] = pd.cut(crimes["Vict Age"], bins=bins, labels=labels, right=True)

# Count number of crimes per age group and sort by age range
victim_ages = crimes["age_group"].value_counts().sort_index()

print(victim_ages)
```
